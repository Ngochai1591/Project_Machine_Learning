```python
def crawl_news_corana(url):
    result = []
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'lxml', from_encoding = 'utf-8')
    
    #find the paragraph
    paragraph = soup.find_all("p")
    
    #print(paragraph)
   
   
    list_para = []
    for text in paragraph:
        #print(text)
        list_para.append(text.text)  
        
    #remove unwanted_text
    
    
    #remove 'For Latest News and Updates'
    #remove link
    #remove which length < 70 char
    unwanted_text = ['Read more here',
                     'Read our analysis here',
                     '/',
                     'Here are the latest updates',
                     '__',
                     'ï»¿',
                     '______________________________________________________________',
                     'For Latest News and Updates',
                     '/',
                     '?',
                     'cookie',
                     'cookies',
                     'settings',
                     'ABP',
                     'Sign up',
                     'inbox',
                     'copyright']
    count = 0
    count_remove_len70 = 0
    
    #replace special character to ''
    for text in list_para:
        text.replace('"', '')
    
    for i in range(len(unwanted_text)):
        count_remove = 0
        for text in list_para:
            if unwanted_text[i].lower() in text.lower() or not text:
                count+=1
                count_remove += 1
                #print(text)
                list_para.remove(text)
            #remove text which length < 70
            elif len(text) < 70:
                list_para.remove(text)
                count_remove_len70 += 1
                count += 1
        print('text =',unwanted_text[i], ', have removed',count_remove)
        
    
    if(count_remove_len70 != 0):
            print('text= len < 70',' have removed', count_remove_len70)
    print('Have removed: ',count)
    print('----------------------------------------------------')
    """
    #print all
    for i in range(len(list_para)):
        print('\n',i,'. ',list_para[i], 'length= ', len(list_para[i]))
    
    
    """
    return list_para
            
    
```

```python
def save_to_file(list_para, redirect):
    #get index of last file
    folder_path = redirect
    dirs = os.listdir(folder_path)
    last_file_name = 0
    if len(dirs) != 0:
        last_file_name = int(dirs[-1].replace('.txt',''))
    else:
        print('Folder is empty')
        print('Create from 1')
    
    print(last_file_name)
    #create name = old_name + 1 
    new_file_index = last_file_name + 1
    new_file_path = ''
    for i in range(len(list_para)):
        new_file_name = new_file_index + i
        if new_file_name >= 100:
            new_file_name = str(new_file_name)
        elif new_file_name >=10:
            new_file_name = '0'+ str(new_file_name)
        elif new_file_name >=1:
            new_file_name = '00' +str(new_file_name)
        new_file_path = folder_path +'\\'+ new_file_name +'.txt'
        #save text to txt file
        with open(new_file_path, 'w', encoding='utf-8') as wf:
            text = list_para[i]
            wf.write(text)
            wf.close()
```

```python
#crawl dynamic web
#using selenium

from selenium import webdriver
import time
```

```python
def scroll_down(browser):
    #get scroll height
    last_height = int(browser.execute_script('return document.body.scrollHeight'))
    print('===============start==================')
    
    count = 0
    
   
       
    start_position = 0
    dest_position = 0
    step = 500
    i = 0
    #Scroll down  
    while True:
        print(i)
        time.sleep(0.5)
        browser.execute_script('window.scrollTo('+str(start_position)+','+str(dest_position)+')')
        start_position = dest_position
        dest_position = i
        new_height = browser.execute_script('return document.body.scrollHeight')
        
        #handle for every 4000 unit
        if i % 4000 == 0:
            if new_height == last_height:
                print('Done')
                break
            else:
                print('Not Yet')
                last_height = new_height
        
        count+=1
        i+=step
    print(count)
   
```
